```python
import numpy as np
#Normal Distribution Function
def zero_truncated_normal():
  while True:
    aa=np.random.normal()
    if aa==0.0:
      print(aa)
      continue
    else:
      return aa

```

```python
#Batch-Size=1
def data_cleaning_split_initialize(X,Y,h):
  X = preprocessing.normalize(X)
  Y=pd.get_dummies(Y)
  X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.3, random_state=42)
  num_input = X_train.shape[1]
  
  num_hidden = h
  num_output = Y_test.shape[1]
  v = np.random.rand(num_input+1,num_hidden)   #(l+1)xm
  w = np.random.rand(num_hidden+1,num_output)  #(m+1)xn

  Y_train=np.array(Y_train)
  Y_test=np.array(Y_test)
  
  K=[zero_truncated_normal(),zero_truncated_normal()]
  return X_train,Y_train,v,w,X_test,Y_test,K
```

```python
def test_prediction(xt,yt,v,w,K):
  crct=0
  y_pred=[]
  Test_Error=0
  for i in range(len(yt)):
    yc=[]
    Ii=xt[i]
    Ii=np.insert(Ii, 0, 1, axis=0)
    Oi=Ii
    Ih=np.matmul(np.transpose(v),Oi)
    Oh=K[0]+K[1]*Ih
  
        
    Oh=np.insert(Oh, 0, 1, axis=0)
    Io=np.matmul(np.transpose(w),Oh)
    Oo=[]
    for io in Io:
      Oo.append(np.exp(io)/np.sum(np.exp(Io)))
    Test_Error+=-sum(yt[i]*np.log2(Oo))
    
      
    yc=[0]*len(Oo)
    yc[np.argmax(Oo)]=1
    y_pred.append(yc)
    if list(yt[i])==yc:
      crct+=1

  return Test_Error,crct,y_pred
```

```python
import math
def forward_backward_softmax_cross_entropy(x,y,v,w,eta,epochs,xt,yt,K):
  kzero=[]
  kone=[]
  xe=[]
  c=0
  ycap=[]
  Train_Loss=[]
  Test_Loss=[]
  Train_Acc=[]
  Test_Acc=[]
  print("Initial Weights between i/p & hidden:",v)
  print("Initial Weights between hidden & o/p:",w)
  print("k0,k1:",K)
    
  #Training Part
  for t in range(epochs):
    Error=0
    c=0
    print("Epochs:-",t)
    for i in range(len(y)):
      #Forward Pass
      Ii=x[i]
      Ii=np.insert(Ii, 0, 1, axis=0)
    
      Oi=Ii
      Ih=np.matmul(np.transpose(v),Oi)
    
      Oh=K[0]+K[1]*Ih
    
      Oh=np.insert(Oh, 0, 1, axis=0)
     
      Io=np.matmul(np.transpose(w),Oh)
     
      Oo=[]
      for io in Io:
        Oo.append(np.exp(io)/np.sum(np.exp(Io)))
    
      Error+=-sum(y[i]*np.log2(Oo))
      yc=[0]*len(Oo)
      yc[np.argmax(Oo)]=1
      if yc==list(y[i]):
        c=c+1
        continue
      else:
        #Backward Pass
        Ebyw1=Oo-y[i]
        Ebyw1=Ebyw1.reshape(-1,1)
        Oh=Oh.reshape(-1,1)
        delEbydelw=np.dot(Oh,np.transpose(Ebyw1))
        delEbydelv=np.dot(w,Ebyw1)
        Ii=Ii.reshape(-1,1)
        delEbydelv=delEbydelv*K[1]
        delEbydelv=np.delete(delEbydelv,0)
        delEbydelv=delEbydelv.reshape(-1,1)
        delEbydelv=np.dot(Ii,np.transpose(delEbydelv))
        delEbydelk0=np.average(np.dot(w,Ebyw1))
        Ih=Ih.reshape(-1,1)
        delEbydelk1=np.dot(np.transpose(Ih),np.delete(np.dot(w,Ebyw1),0))
        w=w-np.dot(eta,delEbydelw)
        v=v-np.dot(eta,delEbydelv)
        K[0]=K[0]-np.dot(eta,delEbydelk0)
        K[1]=K[1]-np.dot(eta,delEbydelk1[0])
    
    print("K's(K0 and K1):-",K)
    #Testing_Part
    Test_Error,crct,y_pred=test_prediction(xt,yt,v,w,K)

    Train_Acc.append(c/len(y))
    Test_Acc.append(crct/len(yt))
    Test_Loss.append(Test_Error)
    Train_Loss.append(Error)
    kzero.append(K[0])
    kone.append(K[1])
    xe.append(t)
    if c==len(y):
      print("Convergence")
      print(c)
      print(len(y))
      break
    else:
      c=0
    # print(Ih,Oh,Io,Oo,delEbydelw,delEbydelv)
    

  print("Total Epochs:",t)
  print("Final Weights between i/p & hidden:",v)
  print("Final Weights between hidden & o/p:",w)
  print("Final k0:",K[0])
  print("Final k1:",K[1])
  print("Final Training Error:",Error)
  
  fig = plt.figure(figsize = (10, 5))
  plt.plot(xe,Train_Loss,color="black")
  plt.xlabel("Number of epochs")
  plt.ylabel("Error Loss Function")
  plt.show()

  fig = plt.figure(figsize = (10, 5))
  plt.plot(xe, Train_Acc, 'b', label='Training_Accuracy')
  plt.plot(xe, Test_Acc, 'y', label='Test_Accuracy')
  plt.title('Train vs Test accuacy wrt epochs')
  plt.xlabel('Epochs')
  plt.ylabel('Train_Test_Accuracy')
  plt.legend()
  plt.show()

  fig = plt.figure(figsize = (10, 5))
  plt.plot(xe, Train_Loss, 'r', label='Training_Loss')
  plt.plot(xe, Test_Loss, 'b', label='Test_Loss')
  plt.title('Train vs Test Loss wrt epochs')
  plt.xlabel('Epochs')
  plt.ylabel('Train_Test_Loss')
  plt.legend()
  plt.show()


  fig = plt.figure(figsize = (10, 5))
  plt.plot(xe, kzero, 'r', label='k0')
  plt.plot(xe, kone, 'b', label='k1')
  plt.title('K0 and K1 changes vs epochs')
  plt.xlabel('Epochs')
  plt.ylabel('K values')
  plt.legend()
  plt.show()

  
  #Final Testing Part
  Test_Error,crct,y_pred=test_prediction(xt,yt,v,w,K)

  print("Final Test Accuracy:",crct/len(yt))
  y_est=y_pred
  y_pred=np.argmax(np.array(y_pred), axis=1)
  y_true=np.argmax(np.array(yt), axis=1)
  print(classification_report(y_true, y_pred))
  print(precision_recall_fscore_support(y_true, y_pred))
  print("Accuracy:",accuracy_score(y_true,y_pred))
  print("Precision:",precision_score(y_true,y_pred,average='weighted'))
  print("Recall:",recall_score(y_true,y_pred,average='weighted'))
  print("F1_score:",f1_score(y_true,y_pred,average='weighted'))
  print("ROC_AUC_Score_array:",roc_auc_score(yt,y_est,average=None))
  print("ROC_AUC_Score_average:",roc_auc_score(yt,y_est,average='weighted'))
```
