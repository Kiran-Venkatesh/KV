```
import requests
from urllib.parse import urlparse, urljoin
from bs4 import BeautifulSoup

def is_valid(url):
    parsed = urlparse(url)
    return bool(parsed.netloc) and bool(parsed.scheme)





def get_all_website_links(mainUrl,thresh):
      url_queue = []
      url_queue.append(mainUrl)
      urls = set()
      c=0
      while len(url_queue) != 0:

        url = url_queue.pop(0)

        domain_name = urlparse(url).netloc
        soup = BeautifulSoup(requests.get(url).content, "html.parser")

        for a_tag in soup.findAll("a"):
                href = a_tag.attrs.get("href")
                if href == "" or href is None:
                    # href empty tag
                    continue
                href = urljoin(url, href)
                parsed_href = urlparse(href)
                href = parsed_href.scheme + "://" + parsed_href.netloc + parsed_href.path
                if not is_valid(href):
                    # not a valid URL
                    continue
                if href in urls:
                    # already in the set
                    continue
                if domain_name not in href:
                    # external link
                    continue
                
                urls.add(href)
                url_queue.append(href)
                c=c+1
                if c%500==0:
                  print(c)
                if c==thresh:
                  return list(urls)            
      
```

***Data Preprocessing Step***

```
from textblob import TextBlob
import re
import string
from nltk.stem import WordNetLemmatizer
from nltk.tokenize import word_tokenize
# Use TextBlob
def textblob_tokenizer(str_input):
    str_input = re.sub(r'\d+','', str_input)
    str_input=str_input.strip()

    blob = TextBlob(str_input.lower())
    tokens = blob.words
    words = [token.stem() for token in tokens if token.isalpha()]
    lemmatizer=WordNetLemmatizer()
    words=[lemmatizer.lemmatize(word) for word in words]
    return words
```

```
from sklearn.feature_extraction.text import TfidfVectorizer,CountVectorizer
import numpy as np

def cosine_similarity(db,new):
  rank_doc=[]
  res_doc=db.copy()
  Docs=res_doc
  vec = TfidfVectorizer(stop_words='english',tokenizer=textblob_tokenizer)
  X= vec.fit_transform(new)
  query = pd.DataFrame(X.toarray(), columns=vec.get_feature_names())

  c=0  
  print(Docs.columns)
  print(query.columns)
  common_col=list(set(Docs.columns) & set(query.columns))
  print(query[common_col].iloc[0].values)
  f=False
  for q in range(len(query)):
    Q=query[common_col].iloc[q].values
    print("Q:",Q)
    for d in range(len(Docs)):
      D=Docs[common_col].iloc[d].values
      Dfull=Docs.iloc[d].values
      dot_product=np.dot(Q,D)
      mag_Q=np.sqrt(Q.dot(Q))
      mag_D=np.sqrt(Dfull.dot(Dfull))
      cos_Q_D=dot_product/(mag_Q*mag_D)
      rank_doc.append(cos_Q_D)
      if dot_product !=0:
        c=c+1
        print("Doc_number:",d)
        print(dot_product,mag_Q,mag_D,cos_Q_D)
  
  rf=pd.DataFrame(rank_doc,columns=["Cosine_value"])
  rf = rf.sort_values(by=['Cosine_value'], ascending=False)
  rf=rf.reset_index()
  rf.columns=["Doc_number","Cosine_value"]
  return rf.head(c)
query=["Logistic Regression"]
top_ranked_doc=cosine_similarity(tfidf_blog_doc,query)
```
